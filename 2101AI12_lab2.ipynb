```python
import cv2
import numpy as np
# from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

def box_filter(image, kernel_size):
    height, width, channels = len(image), len(image[0]), 3
    filtered_image = np.zeros_like(image, dtype=np.uint8)

    half_kernel = kernel_size // 2

    for i in range(half_kernel, height - half_kernel):
        for j in range(half_kernel, width - half_kernel):
            for c in range(channels):
                total_sum = 0
                for ki in range(-half_kernel, half_kernel + 1):
                    for kj in range(-half_kernel, half_kernel + 1):
                        total_sum += image[i + ki, j + kj, c]

                filtered_image[i, j, c] = total_sum // (kernel_size * kernel_size)

    return filtered_image


image_path = '/content/BARB.BMP'
image = cv2.imread(image_path)
noisy_image = image + np.random.normal(0, 25, image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

kernel_size = 8
box_filtered_image_scratch = box_filter(image, kernel_size)
box_filtered_noisy = box_filter(noisy_image, kernel_size)
box_filtered_image_opencv = cv2.blur(image, (kernel_size, kernel_size))


box_filtered_image_scratch_np = np.array(box_filtered_image_scratch, dtype=np.uint8)
box_filtered_image_opencv_np = np.array(box_filtered_image_opencv, dtype=np.uint8)
box_filtered_noisy_scratch_np = np.array(box_filtered_noisy, dtype=np.uint8)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 2)
plt.title('Box Filter (from scratch)')
plt.imshow(cv2.cvtColor(box_filtered_image_scratch_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 3)
plt.title('Box Filter (OpenCV)')
plt.imshow(cv2.cvtColor(box_filtered_image_opencv_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 4)
plt.title('Noisy Image')
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

image_path = '/content/blobs4.png'
image = cv2.imread(image_path)
noisy_image = image + np.random.normal(0, 25, image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

kernel_size = 8
box_filtered_image_scratch = box_filter(image, kernel_size)
box_filtered_noisy = box_filter(noisy_image, kernel_size)
box_filtered_image_opencv = cv2.blur(image, (kernel_size, kernel_size))


box_filtered_image_scratch_np = np.array(box_filtered_image_scratch, dtype=np.uint8)
box_filtered_image_opencv_np = np.array(box_filtered_image_opencv, dtype=np.uint8)
box_filtered_noisy_scratch_np = np.array(box_filtered_noisy, dtype=np.uint8)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 2)
plt.title('Box Filter (from scratch)')
plt.imshow(cv2.cvtColor(box_filtered_image_scratch_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 3)
plt.title('Box Filter (OpenCV)')
plt.imshow(cv2.cvtColor(box_filtered_image_opencv_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 4)
plt.title('Noisy Image')
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

image_path = '/content/boy.jpg'
image = cv2.imread(image_path)
noisy_image = image + np.random.normal(0, 25, image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

kernel_size = 8
box_filtered_image_scratch = box_filter(image, kernel_size)
box_filtered_noisy = box_filter(noisy_image, kernel_size)
box_filtered_image_opencv = cv2.blur(image, (kernel_size, kernel_size))


box_filtered_image_scratch_np = np.array(box_filtered_image_scratch, dtype=np.uint8)
box_filtered_image_opencv_np = np.array(box_filtered_image_opencv, dtype=np.uint8)
box_filtered_noisy_scratch_np = np.array(box_filtered_noisy, dtype=np.uint8)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 2)
plt.title('Box Filter (from scratch)')
plt.imshow(cv2.cvtColor(box_filtered_image_scratch_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 3)
plt.title('Box Filter (OpenCV)')
plt.imshow(cv2.cvtColor(box_filtered_image_opencv_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 4)
plt.title('Noisy Image')
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

image_path = '/content/cameraman0.png'
image = cv2.imread(image_path)
noisy_image = image + np.random.normal(0, 25, image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

kernel_size = 8
box_filtered_image_scratch = box_filter(image, kernel_size)
box_filtered_noisy = box_filter(noisy_image, kernel_size)
box_filtered_image_opencv = cv2.blur(image, (kernel_size, kernel_size))


box_filtered_image_scratch_np = np.array(box_filtered_image_scratch, dtype=np.uint8)
box_filtered_image_opencv_np = np.array(box_filtered_image_opencv, dtype=np.uint8)
box_filtered_noisy_scratch_np = np.array(box_filtered_noisy, dtype=np.uint8)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 2)
plt.title('Box Filter (from scratch)')
plt.imshow(cv2.cvtColor(box_filtered_image_scratch_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 3)
plt.title('Box Filter (OpenCV)')
plt.imshow(cv2.cvtColor(box_filtered_image_opencv_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 4)
plt.title('Noisy Image')
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

image_path = '/content/coins1.png'
image = cv2.imread(image_path)
noisy_image = image + np.random.normal(0, 25, image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

kernel_size = 8
box_filtered_image_scratch = box_filter(image, kernel_size)
box_filtered_noisy = box_filter(noisy_image, kernel_size)
box_filtered_image_opencv = cv2.blur(image, (kernel_size, kernel_size))


box_filtered_image_scratch_np = np.array(box_filtered_image_scratch, dtype=np.uint8)
box_filtered_image_opencv_np = np.array(box_filtered_image_opencv, dtype=np.uint8)
box_filtered_noisy_scratch_np = np.array(box_filtered_noisy, dtype=np.uint8)


plt.figure(figsize=(15, 5))


plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 2)
plt.title('Box Filter (from scratch)')
plt.imshow(cv2.cvtColor(box_filtered_image_scratch_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 3)
plt.title('Box Filter (OpenCV)')
plt.imshow(cv2.cvtColor(box_filtered_image_opencv_np, cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1, 4, 4)
plt.title('Noisy Image')
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def apply_median_filter(image, kernel_size):
    rows, cols, channels = image.shape
    result = np.zeros_like(image, dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            for c in range(channels):
                start_row, end_row = max(0, i - kernel_size // 2), min(rows, i + kernel_size // 2 + 1)
                start_col, end_col = max(0, j - kernel_size // 2), min(cols, j + kernel_size // 2 + 1)

                neighborhood = image[start_row:end_row, start_col:end_col, c]
                result[i, j, c] = np.median(neighborhood)

    return result

image_path = "/content/BARB.BMP"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

noisy_image = original_image + np.random.normal(0, 25, original_image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

median_filtered_image_custom = apply_median_filter(noisy_image, kernel_size=7)

median_filtered_image_opencv = cv2.medianBlur(noisy_image, ksize=7)

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Median Filtered Image of Noisy (Custom)")
plt.imshow(median_filtered_image_custom)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Median Filtered Image of Noisy (OpenCV)")
plt.imshow(median_filtered_image_opencv)
plt.axis('off')

plt.show()

image_path = "/content/blobs4.png"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

noisy_image = original_image + np.random.normal(0, 25, original_image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

median_filtered_image_custom = apply_median_filter(noisy_image, kernel_size=7)

median_filtered_image_opencv = cv2.medianBlur(noisy_image, ksize=7)

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Median Filtered Image of Noisy (Custom)")
plt.imshow(median_filtered_image_custom)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Median Filtered Image of Noisy (OpenCV)")
plt.imshow(median_filtered_image_opencv)
plt.axis('off')

plt.show()

image_path = "/content/boy.jpg"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

noisy_image = original_image + np.random.normal(0, 25, original_image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

median_filtered_image_custom = apply_median_filter(noisy_image, kernel_size=7)

median_filtered_image_opencv = cv2.medianBlur(noisy_image, ksize=7)

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Median Filtered Image of Noisy (Custom)")
plt.imshow(median_filtered_image_custom)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Median Filtered Image of Noisy (OpenCV)")
plt.imshow(median_filtered_image_opencv)
plt.axis('off')

plt.show()

image_path = "/content/cameraman0.png"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

noisy_image = original_image + np.random.normal(0, 25, original_image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

median_filtered_image_custom = apply_median_filter(noisy_image, kernel_size=7)

median_filtered_image_opencv = cv2.medianBlur(noisy_image, ksize=7)

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Median Filtered Image of Noisy (Custom)")
plt.imshow(median_filtered_image_custom)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Median Filtered Image of Noisy (OpenCV)")
plt.imshow(median_filtered_image_opencv)
plt.axis('off')

plt.show()

image_path = "/content/coins1.png"
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

noisy_image = original_image + np.random.normal(0, 25, original_image.shape)
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

median_filtered_image_custom = apply_median_filter(noisy_image, kernel_size=7)

median_filtered_image_opencv = cv2.medianBlur(noisy_image, ksize=7)

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Median Filtered Image of Noisy (Custom)")
plt.imshow(median_filtered_image_custom)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Median Filtered Image of Noisy (OpenCV)")
plt.imshow(median_filtered_image_opencv)
plt.axis('off')

plt.show()
```
